## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
#separando a categoria
officesupplies = df_dsa[df_dsa['Categoria'] == 'Office Supplies']
```

```python
#agrupando por cidade
totalcidades = officesupplies.groupby('Cidade')['Valor_Venda'].sum()

totalcidades.head()
```

```python
# ordenando no sentido descendente
totalcidades.sort_values (ascending = False)
```

```python
#Retorna a cidade com maior valor
resposta1 = totalcidades.idxmax()
print('a cidade com o maior valor de venda para a categoria "Office Suplies" é: ', resposta1)
```

## Pergunta de Negócio 2:

Demonstre o resultado através de um gráfico de barras.

## Pergunta de Negócio 3:

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
# Agrupamento e ordenação dos valores por segmento com maior numero de vendas
analisesegmento = df_dsa.groupby('Segmento')['Valor_Venda'].sum().sort_values(ascending = False).reset_index()

analisesegmento
```

```python
total = analisesegmento['Valor_Venda'].sum()
total
```

```python
analisesegmento['pct'] = ((analisesegmento['Valor_Venda'] / total)*100).round(2)

analisesegmento
```

```python
#Opção 1
plt.figure(figsize = (7,7))
plt.pie(analisesegmento['pct'], 
        labels = analisesegmento['Segmento'],
        autopct = lambda pct: f'{pct:.2f}%'
       )
plt.show()
```

```python
#Opção 2
plt.figure(figsize = (7,7))
plt.pie(analisesegmento['pct'], 
        labels = analisesegmento['Segmento'], 
        autopct = lambda pct: f'{pct:.2f}%'
       )

#Circulo do meio
centre_circle = plt.Circle((0, 0), 0.70, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
df_dsa.head()
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_dsa['Valor_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])
```

```python
df_dsa.head()
```

```python
antes = round(df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda'].mean(),2)
antes
```

```python
depois = round(df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Desconto'].mean(),2)
depois
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
#Obter a informação mês

df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
df_dsa.head()
```

```python
resumo = df_dsa.groupby(['Segmento', 'Ano', 'Mes'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
resumo
```

```python
ano = resumo.index.get_level_values(1)
seg = resumo.index.get_level_values(0)
mes = resumo.index.get_level_values(2)
```

```python
plt.figure(figsize = (12, 6)) #Tamanho da figura
sns.set()
fig1 = sns.relplot(kind = 'line',
                  data = resumo,
                  y = 'mean',
                  x = mes,
                  hue = seg,
                  col = ano,
                  col_wrap = 4)
plt.show
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
df_dsa.head()
```

```python
agru = df_dsa.groupby(['Categoria', 'SubCategoria']).sum('Valor_Venda').sort_values('Valor_Venda', ascending = False).head(12)
agru
```

```python
agru2 = agru[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
agru2
```

```python
# Criamos outro dataframe somente com os totais por categoria
agrucat = agru2.groupby('Categoria').sum(numeric_only = True).reset_index()
agrucat
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Define a função autopct_format
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n(${val:,})'  # Formato personalizado com percentual e valor
    return my_format

# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(agrucat['Valor_Venda'], 
            radius = 1,
            labels = agrucat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(agru2['Valor_Venda'],
            radius = 0.9,
            labels = agru2['SubCategoria'],
            autopct = autopct_format(agru2['Valor_Venda']),
            colors = cores_subcategorias, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(agru2['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
